**Cell 1 — Environment check.** Same as Session 1: confirms your student, batch, and project values are set correctly.

```python
import os

STUDENT_ID = os.environ.get("STUDENT_ID", "NOT_SET")
BATCH_ID = os.environ.get("BATCH_ID", "NOT_SET")
PROJECT_ID = os.environ.get("GOOGLE_CLOUD_PROJECT", "NOT_SET")
REGION = "asia-south1"
GCS_BUCKET = "xro-lab-data"
MY_FOLDER = f"{BATCH_ID}/{STUDENT_ID}/"

print(f"Student ID  : {STUDENT_ID}")
print(f"Batch ID    : {BATCH_ID}")
print(f"GCP Project : {PROJECT_ID}")
print(f"Region      : {REGION} (Mumbai)")
print(f"My GCS path : gs://{GCS_BUCKET}/{MY_FOLDER}")
```

**Cell 2 — Connect to Vertex AI.** This initializes the Vertex AI SDK for your project and region. No API key is needed — the VM's attached service account handles authentication.

```python
import vertexai
from vertexai.generative_models import GenerativeModel
from google.cloud import storage

vertexai.init(project=PROJECT_ID, location=REGION)

print("Connected to Vertex AI")
print(f"Project : {PROJECT_ID}")
print(f"Region  : {REGION}")
```

**Cell 3 — Your first Gemini call.** This sends a real prompt to Gemini 1.5 Flash and gets a response back. Note the token count and response time — both matter once you're paying for this at scale.

```python
import time

model = GenerativeModel("gemini-2.5-flash")

t0 = time.time()
response = model.generate_content(
    "You are an AI engineer at a company in Guwahati, Assam. "
    "In exactly 3 sentences, tell a student from Bodoland who is "
    "learning AI and Google Cloud what excites you most about "
    "building AI systems on GCP."
)
elapsed = round(time.time() - t0, 2)

print(response.text)
print()
print(f"Response time : {elapsed}s")
print(f"Tokens used   : {response.usage_metadata.total_token_count}")
```

**Cell 4 — Read and write to your GCS folder.** Each student has an isolated folder path. This writes a test file and lists what's already there.

```python
gcs = storage.Client(project=PROJECT_ID)
bucket = gcs.bucket(GCS_BUCKET)

test_content = f"XRO Tech | {BATCH_ID} | {STUDENT_ID} | Session 2 check"
blob = bucket.blob(f"{MY_FOLDER}session2_check.txt")
blob.upload_from_string(test_content)

print(f"File saved: {MY_FOLDER}session2_check.txt")
print()

blobs = list(bucket.list_blobs(prefix=MY_FOLDER))
print(f"Files in your folder ({len(blobs)} total):")
for b in blobs:
    fname = b.name.replace(MY_FOLDER, "")
    print(f"  {fname}  ({b.size} bytes)")
```

**Cell 5 — Prompt engineering: try it yourself.** Change `PERSONA` and `QUESTION` below and re-run the cell. Watch how the response changes based on who the model is told to be.

Try changing `PERSONA` to:
- `"You are a recruiter at TCS Guwahati."`
- `"You are a final-year student who just got a job offer."`
- `"You are a technical interviewer at an AI startup."`

```python
PERSONA = "You are a senior cloud architect at Google India."
QUESTION = "What is RAG and why should I learn it for my career in NE India?"

response = model.generate_content(
    f"SYSTEM: {PERSONA}\n\n"
    f"QUESTION: {QUESTION}"
)

print(f"Persona  : {PERSONA}")
print(f"Question : {QUESTION}")
print()
print(response.text)
```
