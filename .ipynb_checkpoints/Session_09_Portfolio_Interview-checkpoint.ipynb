**Cell 3 — Technical interview practice.** These are questions you are likely to actually get asked, with the key points worth hitting in an answer. Pick one, say your answer out loud, time yourself, and aim for under 90 seconds. A friend or classmate listening and giving feedback is more useful than rehearsing alone.

```python
interview_qa = [
    {
        "question": "What is RAG? Why use it instead of fine-tuning?",
        "key_points": [
            "RAG = Retrieval Augmented Generation",
            "Problem it solves: LLMs hallucinate and don't know your private data",
            "How: embed documents, store vectors, retrieve relevant chunks, add to prompt",
            "Why not fine-tune: fine-tuning is expensive, needs GPU time, and bakes data into the model weights",
            "RAG is cheaper, updatable without retraining, and auditable -- you can see what it retrieved",
        ],
    },
    {
        "question": "How does vector search work? How is it different from keyword search?",
        "key_points": [
            "Embedding converts text into a vector of numbers (768 dimensions for text-embedding-004)",
            "Similar meaning produces similar vectors, close together in that high-dimensional space",
            "Cosine similarity measures the angle between two vectors -- 1.0 means identical direction",
            "Keyword search finds exact word matches and misses synonyms or paraphrases",
            "Vector search finds semantic similarity -- \"car\" can match \"automobile\"",
        ],
    },
    {
        "question": "How would you scale this RAG system to a million users?",
        "key_points": [
            "Cloud Run already auto-scales and handles concurrent requests",
            "Bottleneck 1: linear vector search does not scale -- Vertex AI Vector Search handles billion-scale lookups",
            "Bottleneck 2: Gemini API rate limits -- add a queue (Cloud Tasks) and retry logic",
            "Caching: cache embeddings of common questions and cache frequent top-k results",
            "CDN: cache static responses for frequently asked questions",
        ],
    },
    {
        "question": "What is a service account, and how is it different from a user account?",
        "key_points": [
            "A service account is a machine identity, not a human one",
            "This lab's VM authenticates to GCP APIs using a service account, with no password",
            "Authentication uses cryptographic key pairs instead of a password",
            "Should follow least privilege -- only the roles it actually needs",
            "A deployed Cloud Run service typically uses its own service account, separate from the VM's",
        ],
    },
]

print("TECHNICAL INTERVIEW PRACTICE")
for i, qa in enumerate(interview_qa, 1):
    question = qa["question"]
    print(f"\nQ{i}: {question}")
    print("Key points to hit:")
    for point in qa["key_points"]:
        print(f"  - {point}")

print("\nExercise: answer Q1 out loud. Time yourself. Target: 90 seconds.")
```
