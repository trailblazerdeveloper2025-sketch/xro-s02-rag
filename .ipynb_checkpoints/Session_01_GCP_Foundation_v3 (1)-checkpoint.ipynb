**Cell 1 — Environment check.** Confirms your JupyterHub session has the right student, batch, and project values. These are set automatically — you don't need to change anything here.

```python
import os

STUDENT_ID = os.environ.get("STUDENT_ID", "NOT_SET")
BATCH_ID = os.environ.get("BATCH_ID", "NOT_SET")
PROJECT_ID = os.environ.get("GOOGLE_CLOUD_PROJECT", "NOT_SET")
GCS_BUCKET = "xro-lab-data"
MY_FOLDER = f"{BATCH_ID}/{STUDENT_ID}/"

print(f"Student ID  : {STUDENT_ID}")
print(f"Batch ID    : {BATCH_ID}")
print(f"GCP Project : {PROJECT_ID}")
print(f"Region      : asia-south1 (Mumbai)")
print(f"My GCS path : gs://{GCS_BUCKET}/{MY_FOLDER}")

issues = []
if STUDENT_ID == "NOT_SET":
    issues.append("STUDENT_ID not set")
if BATCH_ID == "NOT_SET":
    issues.append("BATCH_ID not set")
if PROJECT_ID == "NOT_SET":
    issues.append("GOOGLE_CLOUD_PROJECT not set")

if issues:
    print("Issue found — tell your instructor:")
    for issue in issues:
        print(f"  - {issue}")
else:
    print("All environment variables look correct. Ready to continue.")
```

**Cell 2 — Imports and authentication.** No API key, no password to type. Your JupyterHub session already has a Google Cloud identity attached to it (a service account), and that identity is shared by everyone in the batch.

```python
from google.cloud import storage
import google.auth
import json
import time

credentials, project = google.auth.default()
identity = getattr(credentials, "service_account_email", "unknown")

print(f"Authenticated as : {identity}")
print(f"Project          : {project}")
```

**Cell 3 — Connect to Cloud Storage.**

```python
gcs_client = storage.Client(project=PROJECT_ID)
bucket = gcs_client.bucket(GCS_BUCKET)
bucket.reload()

print(f"Connected to bucket : {bucket.name}")
print(f"Location             : {bucket.location}")
```

**Cell 4 — Write your first file to the cloud.** This writes directly to Google Cloud Storage, not to the VM's local disk and not to your laptop. It will still be there after you log out.

```python
FILENAME = "session1_hello.txt"
FILE_PATH = MY_FOLDER + FILENAME

FILE_CONTENT = f"""Hello from Google Cloud!

Student:  {STUDENT_ID}
Batch:    {BATCH_ID}
Project:  {PROJECT_ID}
Created:  {time.strftime('%Y-%m-%d %H:%M:%S UTC', time.gmtime())}
Path:     gs://{GCS_BUCKET}/{FILE_PATH}
"""

blob = bucket.blob(FILE_PATH)
blob.upload_from_string(FILE_CONTENT, content_type="text/plain")

print(f"File uploaded: gs://{GCS_BUCKET}/{FILE_PATH}")
print(f"Size: {blob.size} bytes")
```

**Cell 5 — Read the file back.** This proves the file actually persisted on GCS rather than only existing in this Python session. A fresh client is used to simulate reading it from somewhere else entirely.

```python
fresh_client = storage.Client(project=PROJECT_ID)
fresh_bucket = fresh_client.bucket(GCS_BUCKET)
fresh_blob = fresh_bucket.blob(FILE_PATH)

downloaded = fresh_blob.download_as_text()

print(downloaded)
print(f"Confirmed: {len(downloaded)} characters read back from Cloud Storage.")
```

**Cell 6 — List everything in your folder.**

```python
blobs = list(bucket.list_blobs(prefix=MY_FOLDER))

print(f"Contents of gs://{GCS_BUCKET}/{MY_FOLDER}")
for b in blobs:
    name = b.name.replace(MY_FOLDER, "")
    size_kb = round(b.size / 1024, 2) if b.size else 0
    print(f"  {name:<35} {size_kb:>7} KB")

print(f"\nTotal objects: {len(blobs)}")
```

**Cell 7 — How access is organized on a shared VM.** On this lab VM, everyone authenticates as the same service account, so what actually keeps your data separate is the folder path convention (`batch2/your_id/`) rather than per-user IAM roles.

The cell below deliberately lists a different student's folder to make this concrete: you'll be able to see it, because the identity is shared. In a production system, each user would have a distinct identity, and IAM would enforce this boundary at the platform level instead of by convention.

```python
if STUDENT_ID != "s01":
    other_student_folder = f"{BATCH_ID}/s01/"
else:
    other_student_folder = f"{BATCH_ID}/s02/"

other_blobs = list(bucket.list_blobs(prefix=other_student_folder))

print(f"Your identity : {identity}")
print(f"Your folder   : gs://{GCS_BUCKET}/{MY_FOLDER}")
print(f"gs://{GCS_BUCKET}/{other_student_folder} -> {len(other_blobs)} object(s) visible")
```
