**Cell 2 — Write the FastAPI app.** This wraps your RAG pipeline in a typed REST API with three endpoints: a health check, a question-answering endpoint, and a conversation-history lookup. Pydantic models define exactly what shape of data goes in and comes out, which is what powers the automatic API documentation at `/docs`.

```python
os.makedirs("fastapi_app", exist_ok=True)

main_py = '''from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import os
import json
import time
import numpy as np
import vertexai
from vertexai.language_models import TextEmbeddingModel
from vertexai.generative_models import GenerativeModel
from google.cloud import storage, firestore

app = FastAPI(title="XRO RAG API", version="1.0")
app.add_middleware(CORSMiddleware, allow_origins=["*"], allow_methods=["*"], allow_headers=["*"])

PROJECT_ID = os.environ.get("GOOGLE_CLOUD_PROJECT", "xro-lab")
STUDENT_ID = os.environ.get("STUDENT_ID", "student")
BATCH_ID = os.environ.get("BATCH_ID", "batch2")
REGION = "asia-south1"


class AskRequest(BaseModel):
    question: str
    session_id: str = "default"
    top_k: int = 3


class AskResponse(BaseModel):
    answer: str
    session_id: str
    tokens_used: int
    latency_ms: float


_resources = {}


def get_resources():
    if not _resources:
        vertexai.init(project=PROJECT_ID, location=REGION)
        _resources["em"] = TextEmbeddingModel.from_pretrained("text-embedding-004")
        _resources["gm"] = GenerativeModel("gemini-2.5-flash")
        gcs = storage.Client(project=PROJECT_ID)
        blob = gcs.bucket("xro-lab-data").blob(f"{BATCH_ID}/{STUDENT_ID}/session3_vector_index.json")
        _resources["idx"] = json.loads(blob.download_as_text())["chunks"]
    return _resources


def cosine_sim(v1, v2):
    a, b = np.array(v1), np.array(v2)
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))


@app.get("/health")
def health():
    return {"status": "ok", "student": STUDENT_ID}


@app.post("/ask", response_model=AskResponse)
async def ask(req: AskRequest):
    r = get_resources()
    t0 = time.time()
    qv = r["em"].get_embeddings([req.question])[0].values
    chunks = sorted(r["idx"], key=lambda c: cosine_sim(qv, c["vector"]), reverse=True)[:req.top_k]
    ctx = "\\n\\n".join([f"[{i + 1}] {c[\'text\']}" for i, c in enumerate(chunks)])
    prompt = f"Use only this context:\\n{ctx}\\n\\nQ: {req.question}\\nA:"
    resp = r["gm"].generate_content(prompt)
    return AskResponse(
        answer=resp.text,
        session_id=req.session_id,
        tokens_used=resp.usage_metadata.total_token_count,
        latency_ms=round((time.time() - t0) * 1000, 1),
    )


@app.get("/conversations/{session_id}")
def get_conversation(session_id: str):
    db = firestore.Client(project=PROJECT_ID)
    docs = (
        db.collection("conversations")
        .document(f"{STUDENT_ID}_{session_id}")
        .collection("messages")
        .stream()
    )
    return {"messages": [{"role": d.get("role"), "content": d.get("content")} for d in docs]}
'''

with open("fastapi_app/main.py", "w") as f:
    f.write(main_py)

requirements_txt = (
    "fastapi==0.110.0\n"
    "uvicorn[standard]==0.29.0\n"
    "google-cloud-aiplatform==1.48.0\n"
    "google-cloud-storage==2.16.0\n"
    "google-cloud-firestore==2.16.0\n"
    "numpy==1.26.4\n"
)
with open("fastapi_app/requirements.txt", "w") as f:
    f.write(requirements_txt)

dockerfile = (
    "FROM python:3.11-slim\n"
    "WORKDIR /app\n"
    "COPY requirements.txt .\n"
    "RUN pip install --no-cache-dir -r requirements.txt\n"
    "COPY . .\n"
    "EXPOSE 8080\n"
    'CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8080"]\n'
)
with open("fastapi_app/Dockerfile", "w") as f:
    f.write(dockerfile)

print("FastAPI app written: fastapi_app/main.py")
print("Endpoints:")
print("  GET  /health")
print("  POST /ask")
print("  GET  /conversations/{session_id}")
print("  GET  /docs   -- auto-generated API documentation")
```

**Cell 3 — Test it locally.** This starts the API on port 8081, sends it two test requests, then shuts it down. If the model call times out on a busy lab day, that's the embedding/Gemini API being rate-limited under load, not a bug in the API code itself.

```python
import subprocess
import time
import urllib.request
import json as jsonlib

proc = subprocess.Popen(
    ["python", "-m", "uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8081"],
    cwd="fastapi_app",
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL,
)
time.sleep(4)

try:
    with urllib.request.urlopen("http://localhost:8081/health") as r:
        print("GET /health ->", jsonlib.loads(r.read()))
except Exception as e:
    print(f"Health check failed: {e} (server may still be starting, try re-running this cell)")

try:
    data = jsonlib.dumps({"question": "What is RAG?", "session_id": "test1"}).encode()
    req = urllib.request.Request(
        "http://localhost:8081/ask",
        data=data,
        headers={"Content-Type": "application/json"},
        method="POST",
    )
    with urllib.request.urlopen(req, timeout=20) as r:
        resp = jsonlib.loads(r.read())
        print(f"POST /ask -> answer: {resp['answer'][:120]}")
        print(f"           tokens: {resp['tokens_used']} | latency: {resp['latency_ms']}ms")
except Exception as e:
    print(f"Ask endpoint failed: {e}")

proc.terminate()
```

**Cell 4 — How a model chooses to call a function.** This is the core idea behind AI agents: instead of always answering directly, the model can decide a question needs a tool, and tell you which tool and with what arguments.

This cell only shows that decision — it doesn't actually run the function and feed the result back to the model. A complete agent would do that in a loop: call the function, give the result back to the model, let it generate the final answer. That's the natural next step if you want to extend this further.

```python
import vertexai
from vertexai.generative_models import GenerativeModel, Tool, FunctionDeclaration

vertexai.init(project=PROJECT_ID, location=REGION)

search_docs_tool = FunctionDeclaration(
    name="search_documents",
    description="Search the document knowledge base for relevant information",
    parameters={
        "type": "object",
        "properties": {
            "query": {"type": "string", "description": "Search query"},
            "top_k": {"type": "integer", "description": "Number of results", "default": 3},
        },
        "required": ["query"],
    },
)

agent_model = GenerativeModel("gemini-2.5-flash", tools=[Tool(function_declarations=[search_docs_tool])])

user_question = "What does the document say about AI in agriculture?"
response = agent_model.generate_content(user_question)

print("Agent response:")
for part in response.candidates[0].content.parts:
    if hasattr(part, "function_call") and part.function_call.name:
        fc = part.function_call
        print(f"  Model chose to call: {fc.name}({dict(fc.args)})")
        print("  In a complete agent, your code would now run this function and return")
        print("  the result to the model so it can generate a final answer.")
    elif part.text:
        print(f"  Direct answer: {part.text[:200]}")
```
